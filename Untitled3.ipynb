```python
cases = ['Confirmed', 'Deaths', 'Recovered', 'Active']
covid19['Active'] = covid19['Confirmed'] - covid19['Deaths'] - covid19['Recovered']

covid19['Country/Region'] = covid19['Country/Region'].replace('Mainland China', 'China')

covid19[['Province/State']] = covid19[['Province/State']].fillna('')
covid19[cases] = covid19[cases].fillna(0)

covid19['Recovered'] = covid19['Recovered'].astype(int)

covid19.sample(6)
```

```python
ship = covid19[covid19['Province/State'].str.contains('Grand Princess')|covid19['Country/Region'].str.contains('Diamond Princess')]

china = covid19[covid19['Country/Region'] == 'China']
row = covid19[covid19['Country/Region'] != 'China']

covid19late = covid19[covid19['Date'] == max(covid19['Date'])].reset_index()
chinalate = covid19[covid19['Country/Region'] == 'China']
rowlate = covid19[covid19['Country/Region'] != 'China']

covid19_group = covid19late.groupby('Country/Region')['Confirmed', 'Deaths', 'Recovered', 
                                                   'Active'].sum().reset_index()
china_group = chinalate.groupby('Country/Region')['Confirmed', 'Deaths', 'Recovered',
                                                  'Active'].sum().reset_index()

rowlate_group = rowlate.groupby('Country/Region')['Confirmed', 'Deaths', 'Recovered',
                                                  'Active'].sum().reset_index()
                                                 
```

```python
temp = covid19.groupby(['Country/Region', 'Province/State',])['Confirmed', 'Deaths', 'Recovered', 'Active'].max()
```

```python
temp = covid19.groupby('Date')['Confirmed', 'Deaths', 'Recovered',
                               'Active'].sum().reset_index()
temp = temp[temp['Date'] == max(temp['Date'])].reset_index(drop = True)
temp['Global Mortality'] = temp['Deaths']/temp['Confirmed']
temp['Deaths per 100 Confirmed Cases'] = temp['Global Mortality']*100
temp.style.background_gradient(cmap = 'Pastel1')
```
